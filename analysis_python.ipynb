                                       Area plots
If want to find and plot the top 5 nations that contributed to immigration to Canada, we can transpose the data and graph it in a stacked line plot (area plot)

```python
df_canada.sort_values(['Total'], ascending=False, axis=0, inplace=True)

# get the top 5 entries
df_t5 = df_canada.head()

# transpose the dataframe
df_t5 = df_t5[years_list].transpose()

df_t5.head()
# the list below gives us the 5 nations the had most immigrants to Canada
# we see tgar in 1980 for example, India had 8880 immigrants, while Pakistan had 978
```

Stacked plots especially helpful because we can easily recognize nations by color


```python
df_t5.index = df_t5.index.map(int)
df_t5.plot(kind='area',
             stacked=False,
             alpha = 0.50,
             figsize=(20, 10)) 

plt.title('Immigration to Canada by Top 5 nations')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
# Looking at the colors, we can a postive trend for all 5 countriers. 
# China had the most of immigrants of any of those nations by the year 2005, it is roughly 4500
```

                           Graphing horizontally 
We could make a slight change to our graph by adding h to bar (barh) which would graph it in horizontal format

```python
# we can start by creating a new df of let us say 10 countries 
df_canada.sort_values(by='Total', ascending= True, inplace= True)

# make a df of 10 nations 
df_nations10 = df_canada['Total'].tail(10)
print("Here is a list of 10 nations and their total numbe of immigrants in ascending order", df_nations10)
```

```python
# we will now plot the numbers above, note we use a barh, not bar
# also note the y label is for labeling 
df_nations10.plot(kind='barh', figsize=(12, 12), color = 'yellow')
plt.xlabel("numbers of immigrants by nations")
plt.title("top 10 nations with numbers of immigrants to Canada for the years: 1980 - 2013")

# because we want to show numbers on the labels, we can use a for loop with value as following:
for index, value in enumerate(df_nations10):
    label = format(int(value), ',')  
    plt.annotate(label, xy=(value - 45000, index - 0.10), color = 'black')
plt.show()

# we can confirm the numbers above by looking at graph below
# for instance, India showed a number of 691904 above, and we can confirm it below
```

                                         Pie Charts
We can use pie chart to compare the numeric proportions of immigrants by nations 

The pie as a circle in whole is divided based on proportions


We can even compare immigrants in groups based on continents they are from

In this case, by looking, we can see which region or continents had the most immigrants and so on

```python
# Create continents groups by groupby function with the sum method 
df_continents = df_canada.groupby('Continent', axis = 0).sum()

print(type(df_canada.groupby('Continent', axis = 0)))

df_continents.head(3)
```

```python
# we will create a list of colors to point to the continent 
colors_list = ['black', 'red', 'yellow', 'pink', 'green', 'blue']
offset_list = [0.0, 0, 0, 0.1, 0.1, 0.2]

df_continents['Total'].plot(kind='pie',
                            figsize=(19, 10),
                            autopct = '%1.2f%%',
                            startangle = 90,
                            shadow = True,
                            labels = None,
                            pctdistance = 1.12)
plt.title('Numbers of immigrants by continent', y= 1.12, fontsize = 1.14)
plt.axis('equal')
# we can also ad the legend to the chart
plt.legend(labels = df_continents.index, loc = 'upper left', fontsize = 10)
plt.show()

# It is easy to see that Esia had roughly half of all immigrants for all years: 1980 - 2013
# Second place comes the continent of Europ with an overall percentage of 22 percent 
```

                                  Box Plots
If we want to find statistical numbers of distribution of our data, we can use box plots

we can then confirm by using .describe() 

Box plots will give us a summary of the 5 key dimensions: min, max, first qtl, second qtl and third one

```python
# we can make a box plot of India and see the distribution of data
# we can start by creating a df for India 
df_india = df_canada.loc[['India'], years_list].transpose()

df_india.head()
```

To create the boxplot, we can simply pass 'box' to kind 


```python
# boxplot of India data 
df_india.plot(kind='box', figsize = (8, 6))
plt.title("box plot of the data of Indian immigrants to Canda")
plt.ylabel('numbers of immigrants from India')

plt.show()

# We have some interesting numbers here 
#average (median number) appear to be around 21000, lowest numbers appear to be around 3000
# highst numbers of immigrants appear to be around 40000
# around 25% of the time period had annual immigrant numbers of roughly 11000
# 75% (third quartile) of time period, had annual immigrant numbers of around 28000 
```

To confirm and compare our data above, we can use describe() and see the results

```python
df_india.describe()

# min is roughly 4211, maximum appear as 36210
# 25% has a value of 10637, close to what we predicted above (11000)
# 75% has 28699, where we said above to be around 28000
```
